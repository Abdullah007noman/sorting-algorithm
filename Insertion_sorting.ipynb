```python
import time
import random
import matplotlib.pyplot as plt
import psutil
import platform

# Insertion Sort Implementation
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Benchmarking Function
def benchmark_insertion_sort(input_sizes):
    times = []
    for size in input_sizes:
        # Generate a random array of the given size
        arr = [random.randint(0, 1000) for _ in range(size)]

        # Start the timer
        start_time = time.time()

        # Perform the insertion sort
        insertion_sort(arr)

        # End the timer
        end_time = time.time()

        # Calculate the time taken and store it
        times.append(end_time - start_time)

    return times

# Function to Plot Runtime vs Input Size
def plot_runtimes(input_sizes, times):
    plt.plot(input_sizes, times, marker='o', linestyle='-', color='b')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Time (seconds)')
    plt.title('Insertion Sort Runtime vs Input Size')
    plt.grid(True)
    plt.show()

# Function to Get System Information
def get_system_info():
    system_info = {
        "CPU": platform.processor(),
        "CPU Cores": psutil.cpu_count(logical=False),
        "Logical CPUs": psutil.cpu_count(logical=True),
        "RAM": f"{round(psutil.virtual_memory().total / (1024 ** 3), 2)} GB",
        "OS": platform.system(),
        "OS Version": platform.version(),
    }
    return system_info

# Main Function
if __name__ == "__main__":
    # Get system information
    system_info = get_system_info()
    print("System Information:")
    for key, value in system_info.items():
        print(f"{key}: {value}")

    # Define input sizes to test
    input_sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]

    # Benchmark insertion sort for the given input sizes
    times = benchmark_insertion_sort(input_sizes)

    # Print the results
    print("\nBenchmark Results:")
    print("Input Sizes:", input_sizes)
    print("Times (seconds):", times)

    # Plot the runtime vs input size
    plot_runtimes(input_sizes, times)
```
